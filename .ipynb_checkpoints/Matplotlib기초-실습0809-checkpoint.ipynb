### **그래프 저장하기 ②** 

```python
import matplotlib.pyplot as plt
import numpy as np
# sin 그래프
x = np.linspace(0,np.pi * 10, 500) #0~31.4 까지 500개의 점을 균일하게 찍기
fig,axes = plt.subplots(2,1) #2행 1열 그래프를 만들어라
axes[0].plot(x, np.sin(x)) #첫 번째 그래프는 사인(sin)그래프
axes[1].plot(x, np.cos(x)) #두 번째 그래프는 코사인(cos)그래프
plt.savefig('sin_cos.png')
```

### **선 그래프 그리기 ①**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-9,10)
y = x **2
plt.plot(x,y,linestyle = "--",marker = 'o' )
# linestyle = ":" , "-" "--"// marker = " * " , " + " , " ^ " , " o "
plt.show()
```

### **선 그래프 그리기 ②**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-9,10)
y1 = x **2
y2 = -x
plt.plot(x,y1, linestyle = "-",marker = "*" , color = "red" , label = "y=x*x")
plt.plot(x,y2, linestyle = ":",marker = "o" , color = "blue" , label = "y=-x")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(shadow = True ,borderpad = 1) #범례 . 위치값도 줄 수 있음
plt.show()
```





## Matplotlib 라이브러리의 다루기






### **선 그래프 그리기**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-9,10)
y1 = x **2
y2 = -x
plt.plot(x,y1, linestyle = "-",marker = "*" , markersize = "8" , markerfacecolor = "blue" , markeredgecolor = "green" , color = "red" , label = "y=x*x")
plt.plot(x,y2, linestyle = ":",marker = "o" , markersize = "20" , markerfacecolor = "red" , markeredgecolor = "yellow" ,color = "blue" , label = "y=-x")
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(shadow = True ,borderpad = 1,loc = "upper center") #범례 . 위치(loc)는 총 9군데 지정할 수 있음
#loc = lower(left , center , right) , upper (@,@,@) , center (@,@) ,  center 단일
plt.show()
```

### **막대 그래프 그리기**



```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-9,10)
plt.bar(x, x**2)
plt.show()
```

### **누적 막대 그래프 그리기**



```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(10) # 아래 막대
y = np.random.rand(10) # 중간 막대
z = np.random.rand(10) # 상단 막대
data = [x,y,z]
x_array = np.arange(10)
for i in range(0,3):
    plt.bar(x_array, data[i],
    bottom = np.sum(data[:i],axis = 0))
plt.show()
```

### **스캐터 그래프 그리기**



```python
#스캐터(산포도)
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.randint(0,1000,100)
sizes = np.pi * 1000 * np.random.rand(100)
plt.scatter(x,y, c = colors, s = sizes, alpha = 0.7)
#alpha = 투명도 
plt.show()
```
